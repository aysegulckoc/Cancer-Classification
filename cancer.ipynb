## Outlier Detection

Veri setinde icindeki aykiri degerlerdir. Ayiklanmazsa model yanlis egitebilir.
Cesitli yontemler vardir. 
Density based ODS --> 
Local outlier factor LOF: bir noktanin local density ile K NN locak density ile karsilastirilmasi
LOFa >1 outlier , LOFa <1 normal 

LRD = Local Reachability Density 
*(en yakin iki komsu--> LRDb LRDd)
LOFa = LRDb + LRDd / LRDa x (1/K=2)

ARD = Average Reachability Distance 
LRD = 1 / ARD

RD = Reachability Distance

```python
y = data.target # target variable
x = data.drop(['target'], axis= 1) # feature
columns = x.columns.tolist()
```

```python
x
```

```python
y
```

```python
clf = LocalOutlierFactor()
y_pred = clf.fit_predict(x)
```

```python
y_pred
```

```python
X_score = clf.negative_outlier_factor_
outlier_score = pd.DataFrame()
outlier_score['score'] = X_score
```

```python
X_score
```

```python
threshold = -2.5
filtre = outlier_score['score'] < threshold 
outlier_index = outlier_score[filtre].index.tolist()
```

```python
plt.figure()
plt.scatter(x.iloc[outlier_index,0],x.iloc[outlier_index,1],color='blue',s=50,label='data point')

radius = (X_score.max() - X_score)/(X_score.max() - X_score.min()) #normalize islemi
# outlier_score['radius']

plt.scatter(x.iloc[:,0],x.iloc[:,1],s=1000*radius,edgecolors='r',facecolors='none',label='data point')
plt.legend()
plt.show()
```

```python
#drop outliers
x = x.drop(outlier_index)
y = y.drop(outlier_index).values
```
